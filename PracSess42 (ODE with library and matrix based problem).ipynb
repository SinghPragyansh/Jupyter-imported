```python
# Que: dS(t)/dt = -S(t), plot approximate nad exact solution.
# Given: S0 = 1, exact solution is : S(t) = exp(-t), interval  = [0,1]
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
f = lambda t, s : -s
```

```python
t_eval = np.arange(0, 1.01, 0.01)
sol = solve_ivp(f, [0,1], [1], t_eval = t_eval)
```

```python
plt.figure(figsize = (12, 4))
plt.subplot(121)
plt.plot(sol.t, sol.y[0])
plt.xlabel("t")
plt.ylabel("S(t)")
plt.subplot(122)
plt.plot(sol.t, sol.y[0] - np.exp(-sol.t))
plt.xlabel("t")
plt.ylabel("S(t) - exp(-t)")
plt.tight_layout()
plt.show()
```

# DE function involving matrix and vector

Que: Let S(t) = column([x(t), y(t)])      
and evolution is given by ODE: 2 x 2 matrix::                 
dS/dt = mat ( [ [0, t^2], [-t, 0] ] )


```python
f = lambda t, s: np.dot(np.array([[0, t**2], [-t, 0]]), s)
t0 = 0
tf = 10
h = 0.01
t = np.arange(t0, tf + h , h)
s0 = np.array([1,1])
sol_array = solve_ivp(f, [t0, tf], s0, t_eval = t)
```

1. x(t) and y(t) are the components of the time dependent position vector
2. The solved output contains the information of both co-ordinates x and y.
3. Both co-ordinate being time dependant, will take their own course with time.
4. Info of y and x coordinate is stored in the 1st and 2nd column respectively

```python
plt.figure(figsize = (8,8))

# y(t) vs x(t)
plt.subplot(311)
plt.plot(sol_array.y.T[:,1], sol_array.y.T[:, 0])
plt.xlabel("x(t)")
plt.ylabel("y(t)")
plt.title(" y(t) vs x(t)")

# y(t) vs t plot
plt.subplot(312)
plt.plot(sol_array.t, sol_array.y[0])
plt.ylabel("y(t)")
plt.xlabel("time")
plt.title(" y(t) vs time ")

# x(t) vs t plot
plt.subplot(313)
plt.plot(sol_array.t, sol_array.y[1])
plt.ylabel("x(t)")
plt.xlabel("time")
plt.title(" x(t) vs time")
```
