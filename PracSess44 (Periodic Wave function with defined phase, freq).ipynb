```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# sampling rate
sr = 100.0

# sample numbers in a second (grid points)
s = 1.0 / sr
t = np.arange(0,1, s)
```

```python
def sin_wave(freq_mu, phase,amplitude, time):
   
    omega = 2*np.pi*freq_mu
    y = amplitude*np.sin(omega*time + phase)
    
    return y
```

```python
freq_mu_1 = 5
phase_1 = 0
amplitude_1 = 1
wave_1 = sin_wave(freq_mu_1, phase_1, amplitude_1, t)
```

```python
plt.plot(t, wave_1, "b")
```

```python
freq_mu_2 = 10
phase_2 = 0
amplitude_2 = 1
wave_2 = sin_wave(freq_mu_2, phase_2, amplitude_2, t)
plt.plot(t, wave_2, "b")
```
