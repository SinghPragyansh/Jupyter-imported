Clearly, it is not useful to express functions as infinite sums because we cannot compute them. That said, it is often useful to approximate functions by using an N th order Taylor series approximation of a function, which is a truncation of its Taylor expansion at some n = N . This technique is especially powerful especially when there is a point around which we have knowledge about a function and all its derivatives.

```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
u = np.linspace(-np.pi, np.pi, 200)
v = np.zeros(len(u))
t = int(input("Numbers of terms in the series: "))
labels = ["First Term", "Second Term", "Third Term", "Fourth Term", "Fifth Term"]
```

```python
for n, label in zip(range(t), labels):
    v = v + (((-1)**n)*(x)**(2*n+1)/np.math.factorial(2*n+1))
    plt.plot(u, v, label = label)


plt.plot(u, np.sin(x), "k", label = "Analytic")
plt.grid()
plt.title("Taylor Series app")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
# Without including label, we can unpack it to as much terms as we like
```

```python
def Sin_fun(x, t):
    y = np.zeros(len(x))
    for n in range(t):
        y = y + (((-1)**n)*(x)**(2*n+1)/np.math.factorial(2*n+1))
    return y
```

```python
# Plotting for any nth order
y_val = Sin_fun(u, t)
plt.plot(u, y_val)
plt.plot(u, np.sin(x), "k")
plt.grid()
plt.title("Taylor Series app")
```
