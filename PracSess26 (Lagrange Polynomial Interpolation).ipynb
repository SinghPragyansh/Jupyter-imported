```python
# Ques: Find the Lagrange basis polynomials for the dataset x = [0, 1, 2] and y = [1, 3, 2]. Ploteach polynomial
```

```python
import numpy as np
import numpy.polynomial.polynomial as poly
import matplotlib.pyplot as plt
plt.style.use("seaborn-poster")
```

```python
u = [0, 1, 2]
v = [1, 3, 2]
coeff_1 = [1,-1.5,.5]
coeff_2 = [0, 2,-1]
coeff_3 = [0,-.5,.5]
```

```python
# Defining the polynomial according to the coeff 
P1_x = poly.Polynomial(coeff_1)
P2_x = poly.Polynomial(coeff_2)
P3_x = poly.Polynomial(coeff_3)
```

```python
# Defining the x grid through list for the polynomial to run over it
x_grid = np.arange(-2, 4.1, 0.1)
```

```python
# access the grid array through the poly func 
P1 = P1_x(x_grid)
P2 = P2_x(x_grid)
P3 = P3_x(x_grid)
```

```python
fig = plt.figure(figsize = (6,4))
plt.plot(x_grid, P1, "b", label = "P1")
plt.plot(x_grid, P2, "r", label = "P2")
plt.plot(x_grid, P3, "g", label = "P3")
plt.title("Lagrange polynomial function")
plt.grid()
```

```python
# Now we get the lagrange equation from these polynomials:
# L(x) = sigma[yi * Pi(x)]
```

```python
# Writing the legendre polynomila function according to the y values 
L_x = 1*P1_x + 3*P2_x + 2*P3_x
L = L_x(x_grid) 
```

```python
plt.figure(figsize = (6,4))
plt.plot(x_grid, L, "b", u, v, "ro")
plt.grid()
```

```python
# Cheching the legendre polynomial interpolation for random x within the grid
interpolated = L_x(3)
print(interpolated)
plt.figure(figsize = (6,4))
plt.plot(x_grid, L, "b", 3, interpolated, "ro")
```
