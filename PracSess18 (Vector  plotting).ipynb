```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use("seaborn-poster")
%matplotlib inline
```

```python
# x: column vector
# b = Ax , where A matrix acts like an operator
# Objective is to plot x and b vector
# Important to note is: the vector plotting consists of column vectors
# The eigen vector is already column vector where as matrix M after
# multiplication with eigen vect, also becomes column vector
# Therefore, fulfilling the criteria of vector plotting 

def plot_vec(x, b, xlim, ylim):
    plt.figure(figsize =(15,9))

    # first two parameters are for origin of arrow
     # 2nd and 3rd parameter is for arrow to read the index of eigen vector
    plt.quiver(0,0,x[0],x[1],x[2], color ="g", label="Original vector")
    plt.quiver(0,0,b[0],b[1],b[2],\
        color="r",angles="xy",\
        scale_units="xy",scale=1,\
        label ="M.x vector")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel("X")
    plt.ylabel("Y")
    plt.legend()
    plt.show()
    
    
M = np.array([[2,0,1],[2,1,5],[3,2,2]])
x = np.array([[1],[1],[1]])
b = np.dot(M, x)
plot_vec(x, b, (0,10), (-50,50))
```

```python
N = np.array([[2,0,1],[2,1,5],[3,4,1]])
t = np.array([[2],[0],[2]])
c = np.dot(N, x)
plot_vec(t, c, (0,10), (-50,50))
```
